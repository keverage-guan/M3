```python
import math
import numpy as np
import statsmodels.api as sm
from statsmodels.base.model import GenericLikelihoodModel
from scipy.stats import poisson
from patsy import dmatrices
import statsmodels.graphics.tsaplots as tsa
from matplotlib import pyplot as plt
from statsmodels.tools.numdiff import approx_hess1, approx_hess2, approx_hess3
import pandas as pd
```

```python
df = pd.read_excel('wildfires.xlsx', sheet_name='Acadia NP')[['CalendarYear', 'SizeClass']]
totalCount = df.groupby(['CalendarYear'])['CalendarYear'].count()
totalCount.head()
```

```python
#Plot the number of strikes starting each month
plt.xlabel('Year')
plt.ylabel('Number of Wildfires')
totalCount.plot()
plt.show()
```

```python
tsa.plot_acf(totalCount, alpha=0.5)
plt.show()
```
